```python
#importing pandas to work with the csv file
import pandas as pd
```

```python
#importing training dataset
df = pd.read_csv("train.csv")
```

```python
#checking the dataframe
df
```

```python
#pulling how many of the passengers were each sex
df.groupby("Sex").size()
```

```python
#pulling how many of the passengers survived
df.groupby("Survived").size()
```

```python
#plotting totals by sex
df.groupby("Sex").size().plot.bar()
```

```python
#plotting survival by sex
df.groupby(["Sex","Survived"]).size().unstack(fill_value=0).plot.bar()
```

```python
#calculating average fare by sex
df.groupby("Sex").mean()
```
